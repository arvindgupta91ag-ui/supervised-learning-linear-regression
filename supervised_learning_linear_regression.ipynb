2.  What are the key assumptions of Simple Linear Regression?
- Key Assumptions of Simple Linear Regression (SLR)
For SLR to give valid and reliable results, certain assumptions must hold true:

Linearity

The relationship between the independent variable (X) and dependent variable (Y) is linear.

Example: As study hours increase, exam scores increase in a straight‑line pattern.

Independence of Errors

Residuals (errors) are independent of each other.

No autocorrelation (important in time‑series data).
Homoscedasticity (Constant Variance)

The variance of residuals is constant across all levels of X.

In other words, the spread of errors should not increase or decrease with X.

Normality of Errors

Residuals should be approximately normally distributed.

This is crucial for hypothesis testing and confidence intervals.

No Perfect Multicollinearity

In SLR, since there’s only one predictor, this means X should not be a constant or perfectly correlated with Y.

In multiple regression, predictors must not be perfectly correlated with each other.
Why These Assumptions Matter
They ensure that the regression line is the best unbiased estimator of the relationship.

Violations (like heteroscedasticity or non‑linearity) can lead to biased coefficients, unreliable predictions, and invalid statistical tests.

9.  Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)

```python
# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample dataset (X = study hours, Y = exam scores)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # Independent variable
Y = np.array([20, 40, 60, 80, 100])            # Dependent variable

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Print slope (coefficient) and intercept
print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)

# Explanation
# Slope (β1 = 20.0): For each additional study hour, the score increases by 20 points.

# Intercept (β0 = 0.0): When study hours = 0, the predicted score is 0.
```
